* 도준은 100, 50, 30의 재산을 가지고 있다. 리스트와 numpy array로 저장하고 데이터타입을 출력하여 비교

```python
dojun_list = [100, 50, 30]
# dojun_arr = np.array((100, 50, 30))
dojun_arr = np.array(dojun_list)
print(type(dojun_list))
print(type(dojun_arr))
```

* 도준의 재산을 100, 50, 30<br>
성준의 재산을 200, 30, 80<br>
이라고 리스트로 저장한 후 더하기 연산하여 결과를 출력

```python
dojun_list = [100, 50, 30]
sungjun_list = [200, 30, 80]
total_list = dojun_list + sungjun_list
print(total_list)
```

* 위 내용을 리스트 대신 numpy array로 저장하여 결과를 출력

```python
dojun_arr = np.array(dojun_list)
sungjun_arr = np.array(sungjun_list)
total_arr = dojun_arr + sungjun_arr
print(total_arr)
```

* 위위의 리스트에 곱하기 연산을 하여 결과를 출력

* 위의 array에 곱하기 연산을 하여 결과를 출력

```python
mult_arr = dojun_arr * sungjun_arr
print(mult_arr)
```

* 도준은 성준에 비해 몇%의 재산을 가지고 있는지 계산하고 출력. 위의 list 및 array는 2020년, 2021년, 2022년도의 재산을 나타냄
